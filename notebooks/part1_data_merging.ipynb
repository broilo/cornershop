# Merging Datasets
* Merge 1: mg1 $\to$ orders \& shoppers
* Merge 2: mg2 $\to$ mg1 \& storebranch
* Merge 3: mg3 $\to$ mg2 \& order_products

## Merge 1: mg1 $\to$ orders \& shoppers
* **order_id**: primary key
* **shopper_id**: foreign key

```python
mg1 =  orders.merge(
    shopper,
    how='left',
    left_on=['shopper_id'],
    right_on=['shopper_id'])

print(mg1.shape)
mg1.head()
```

## Merge 2: mg2 $\to$ mg1 \& storebranch
* **order_id**: primary key
* **store_branch_id**: foreign key

```python
mg2 = mg1.merge(
    storebranch,
    how='left',
    left_on=['store_branch_id'],
    right_on=['store_branch_id'])

mg2.rename({
    'lat_x':'lat_order'
    ,'lng_x':'lng_order'
    ,'lat_y':'lat_storebranch'
    ,'lng_y':'lng_storebranch'
}, axis=1, inplace=True)

print(mg2.shape)
mg2.head()
```

## Merge 3: mg3 $\to$ mg2 \& order_products
* **order_id**: primary key
* **order_id**: foreign key

```python
mg3 = mg2.merge(
    order_products,
    how='left',
    left_on=['order_id'],
    right_on=['order_id'])

print(mg3.shape)
mg3.head()
```

**Notice that:** 
* There are order_id in orders.csv that aren't in order_products.csv
    1. Therefore, to take into account all order_id, *i.e* all buying/deliver orders the correct join must be a left one.
* The four datasets were reduced to only one, *a.k.a.* mg3

# Splitting the dataset
* Dataset to predict: Simulating the real world
    * The orders associated with missing total_minutes values
* Dataset to Train and Test
    * The orders associated with non-missing total_minutes values

**Remember that:** The rows not containing a total_minutes value should be set aside.

## Dataset to predict: Simulating the real world

### The orders associated with missing total_minutes values

```python
df_to_predict = mg3[mg3.total_minutes.isnull()==True]
print(df_to_predict.shape)
df_to_predict.head()
```

## Dataset to train and test

### The orders associated with non-missing total_minutes values

```python
df_tt = mg3[~mg3.total_minutes.isnull()==True]
print(df_tt.shape)
df_tt.head()
```
